```python
import matplotlib.pyplot as plt
```

```python
# GRAFICO DE LINHAS
plt.plot([1.3,2.5,3.6,4.6,5.5], [10,20,30,40,50], label='carro 1')
plt.plot([1.3,2.5,3.6,4.6,5.5], [15,25,35,45,60], label='carro 2')
plt.ylabel('Velocidade')
plt.xlabel('Tempo')
plt.title('Velocidade e tempo')
plt.legend()
plt.show()
```

```python
# GRAFICO DE BARRAS

x = [1,2,3,4] # quantidade de barras
y = [5,10,15,20] # valores das barras
plt.bar(x, y)
plt.show()
```

```python
# GRAFICO DE DISPERSS√ÉO

xp1 = [1, 2, 3, 4, 5, 6, 7] # quantidade de pontos
yp1 = [5, 10, 3, 6, 7, 8, 10] # valores
xp2 = [2, 4, 6, 3, 7, 8, 6] # quantidade de pontos
yp2 = [3, 1, 2, 2, 3, 4, 8] # valores
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 5, 6, 7, 8]
plt.plot(x,y)
plt.scatter(xp1,yp1)
plt.scatter(xp2,yp2)
```

```python
# GRAFICO DE PIZZA

label = ['floresta', 'deserto']
dados = [35, 35] # completa os dados para que fiquem equivalentes a 100%
fig, ax = plt.subplots() 

ax.pie(dados, labels = label, autopct = '%1.1f%%')
```
